
## Contexte
"swipe right" lorsque l'utilisateur aime le profil visionné
"swipe left" lorsque l'utilisateur n'aime pas le profil visionné

On cherche à comprendre via ce dataset ce qui rend les gens intéréssés les uns des autres
Ce dataset est basé sur une expérience de "speed dating"
Chaque ligne représente un "speed dating" et indique si chacunes des deux personnes accepterait un 2ème RDV


## Objectifs
comprendre ce qui intéresse les gens les uns par rapport aux autres pour aller au 2ème RDV

## Déroulé de l'experience
chaque participant a un 1er RDV de 4 minutes avec tous les autres participants du sexe opposé
A la fin des 4 minutes, on demande aux participants s'ils souhaitent aller au second RDV ET ils notent leur date sur 6 attributs (critères):
- attractivité = attr
- sincérité = sinc
- intelligence = intel
- fun = fun
- ambition = amb
- intérets communs = shar


NOTES D'ATTRACTIVITE:
- attr1_1: en quoi l'attirance physique est importante pour vous dans un RDV
- attr: à quel point vous avez trouvé attirant votre date
- attr_0: à quel point votre date vous a trouvé attirant

NOTES DE SINCERITE:
- sinc1_1: en quoi la sincérité est importante pour vous dans un RDV
- sinc: à quel point vous avez trouvé votre date sincère
- sinc_0: à quel point votre date vous a trouvé sincère

...

## Idées d'exploration
- Quels sont les attributs les moins souhaités chez un partenaire masculin 
- Quels sont les attributs les moins souhaités chez un partenaire féminin
- Sont-ils différents
- Les intérêts communs sont-ils plus importants que la race de chaque participants ?
- Pour obtenir un second RDV, est-il préférable de faire le 1er RDV en début de soirée ou en fin de soirée ?

### Ce que les utilisateurs de tinder cherchent en utilisant l'application

```python

# Le but du speed dating pour les participants
display(df['goal'].value_counts().reset_index())
fig = px.histogram(df, x='goal',nbins=15, category_orders=dict({'goal':['Get a fun night', 'Meet new people', 'Get a date', 'To say "I did it"', 'Other', 'Looking for serious relationship']}))
fig.show()
```

```python
df["goal"] #type float, je veux des type str pour pouvoir changer les valeurs pour que ce soit plus explicite
```

```python
df['goal'] = df['goal'].astype(str)
```

```python
df['goal'] = df['goal'].replace({'1.0':'Get a fun night', '2.0':'Meet people', '3.0':'Get a date', '4.0':'Looking serious relationship', '5.0':'To say i did it', '6.0':'Other'})
```

```python
df['goal']
```

```python
display(df['goal'].value_counts().reset_index())
fig = px.histogram(df, x='goal',nbins=15, title="What users are looking for", color='goal')
```

```python
fig.update_layout(showlegend=False)
fig.show()
```
